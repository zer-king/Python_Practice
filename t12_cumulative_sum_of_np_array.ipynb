```python
import pandas as pd
import numpy as np
import random
# t12: Calculate the cumulative sum of a NumPy array and store the results in a new Pandas DataFrame column.
df=pd.read_csv('employee_performance_data.csv')
# Generate a NumPy array of random integers representing training hours
# the np.array has to have the random and for loop inside [] within the np.array() function to generate the array of random integers as without the [] it will generate only one random integer and np.array needs a 2d input to create an array of multiple values.
# the reason it has to be within [] is that the [] creates a list comprehension which generates a list of random integers for each iteration of the for loop, and then np.array() converts that list into a NumPy array.
# as without the [] it will generate a single random integer for each iteration of the for loop and np.array() will only convert that single integer into a NumPy array, resulting in an array with only one element instead of multiple elements and so an invalid input for np.array() which expects a 2D input.
np_array=np.array([random.randint(1,100) for i in range(len(df))])
np_array=np.array(random.randint(1,100) for i in range(len(df)))
pd_array=pd.DataFrame(np_array,columns=['Training_Hours'])
pd_array['Cumulative_Sum']=np.cumsum(pd_array['Training_Hours'])

np_array
cumlist=[]
total=0
for i in np_array:
    # total=total+i
    total+=int(i)
    cumlist.append(total)
print("cumlist:\n",cumlist)

print(np_array)
print(pd_array)
print(df.columns)
df['Cumulative_Sum'] = np.cumsum(df['Training_Hours'])
print(df)
```
