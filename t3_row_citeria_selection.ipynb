```python
import pandas as pd
import numpy as np
import random

#set random seed for reproducibility
np.random.seed(40)

# Parameters for random DataFrame
num_rows = 50
projects = ["Project A", "Project B", "Project C", "Project D", "Project E"]
phases = ["Planning","Development", "Testing", "Deployment"]
categories = ["Labour", "Materials", "Equipment", "Miscellaneous"]
locations = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]
date_range = pd.date_range(start='2025-01-01', end='2025-11-22', freq='D')
flattened_date_range = date_range.tolist()


#print(date_range)
# generate random data
data = {
    "Project": np.random.choice(projects, num_rows),
    "Phase": np.random.choice(phases, num_rows),
    "Category": np.random.choice(categories, num_rows),
    "Location": np.random.choice(locations, num_rows),  
    "Date": np.random.choice(flattened_date_range, num_rows),
    "Cost": np.round(np.random.uniform(1000, 10000, num_rows), 0)
}

df = pd.DataFrame(data)
df.to_csv('construction_data.csv', index=False)
dataHead = df.head(10)

#print(dataHead)
# This will tell which rows have project as "Project A" and will return a boolean series
selected_rows=df["Project"]=="Project A"
# Typically we need the exact data, so we use this boolean series to filter the dataframe
selected_rows=df[selected_rows]
# or simply all in one line
selected_rows=df[df["Project"]=="Project A"]
print(selected_rows)
```
