```python
import pandas as pd
import numpy as np  

# t11: Correlation Matrix Calculation for Construction Project Data
# Load the dataset#
df=pd.read_csv('construction_data.csv')
# Calculate the correlation matrix, however, this will not work if there are non-numeric columns like Project Name#
# So we need to select only numeric columns#
# First, let's inspect the dataframe to see its structure#
#print(df.columns)
# Check the data types of each column#
#print(df.dtypes)
# Convert non-numeric columns to numeric if necessary#
# For example, if 'Project' is a categorical variable, we can convert it to numeric codes #  
df['Project_Codes']=df['Project'].astype('category').cat.codes
# Similarly for 'Phase' column
df['Phase_Codes']=df['Phase'].astype('category').cat.codes
# Now select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=[np.number])
# Calculate the correlation matrix
correlation_matrix=numeric_df.corr()
# Print the correlation matrix#
# You can also use help to understand the select_dtypes function #
#help( df.select_dtypes)
#print(correlation_matrix)

# T11 Further practice task: 
# Create a DataFrame for Employee Performance Data, 
# check data type of each column, 
# convert categorical columns to numeric codes, 
# and calculate the correlation matrix.:
import random
random.seed(42)

Employee_Name = ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack']
Department = ['HR', 'Finance', 'IT', 'Marketing', 'Sales']
Years_of_Experience = [random.randint(1,20) for i in range(len(Employee_Name))]
Training_Hours = [random.randint(10,100) for i in range(len(Employee_Name))]
Performace_Score = [random.randint(1,10) for i in range(len(Employee_Name))]    
Promotion_Options = ['Promoted', 'Not Promoted']
Promotion_Status = [random.choice(Promotion_Options) for i in range(len(Employee_Name)) ]

Employee_Performance_df = pd.DataFrame({
    'Employee_Name': Employee_Name,
    'Department' : random.choices(Department, k=len(Employee_Name)),
    'Years_of_Experience': Years_of_Experience,
    'Training_Hours': Training_Hours,   
    'Performace_Score': Performace_Score,
    'Promotion_Status': Promotion_Status

})

#print(Employee_Performace_df)
print(Employee_Performance_df.dtypes)
# Convert categorical columns to numeric codes
Employee_Performance_df["Department_Codes"]=Employee_Performance_df['Department'].astype('category').cat.codes
Employee_Performance_df["Department_Codes"]=Employee_Performance_df['Promotion_Status'].astype('category').cat.codes
# Now select only numeric columns for correlation calculation
Employee_Performance_df_numeric= Employee_Performance_df.select_dtypes(include='number')
# Calculate the correlation matrix
correlation_matrix_emp= Employee_Performance_df_numeric.corr()
print(correlation_matrix_emp)


```
