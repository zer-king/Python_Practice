```python
#make a shops data file
import pandas as pd
import numpy as np
import random
# Set random seed for reproducibility
np.random.seed(42)
# Generate random data
# Create a date range
dates_range=pd.date_range(start='2025-01-01', end='2025-11-22',freq='D')
# Define shops and item types
shops=['Shop_A','Shop_B','Shop_C','Shop_D','Shop_E']
types_of_items=['Electronics','Clothing','Groceries','Furniture','Toys']
# Number of rows to generate
num_rows=100
# Create the DataFrame with random data for each column
data ={
    'Date' : np.random.choice(dates_range,num_rows),
    'Shop' : np.random.choice(shops,num_rows),
    'Item_Type' : np.random.choice(types_of_items,num_rows),
    'Quantity_Sold' : np.random.randint(1,20,num_rows),
    'Price_Per_Item' : np.round(np.random.uniform(5.0,500.0,num_rows),2)
}
# Create DataFrame and save to CSV
dataframe=pd.DataFrame(data)
# set index as False to avoid writing row numbers into the CSV file
dataframe.to_csv('shops_data.csv',index=False)
# Display first and last 7 rows of the DataFrame
dfHead=dataframe.head(7)
# Display last 7 rows of the DataFrame
dfTail=dataframe.tail(7)
# Print the results to the console for verification with spacing with \n
print("dfHead:",dfHead,"\n \n")
print("dfTail:",dfTail)
```
